```python
filename = '../resources/DataSujetos/N1_suj1.csv'
df = pd.read_csv(filename, header=None)
n = df.shape[0]
aristas_posibles = (n*n-1)//2
df -= np.diag(np.ones(n))
filename = '../resources/DataSujetos/W_suj1.csv'
df_w = pd.read_csv(filename, header=None)
df_w -= np.diag(np.ones(n))
```

```python
sort_corrs = sorted(df.values.flatten(),reverse=True)
sort_corrs_w = sorted(df_w.values.flatten(),reverse=True)
modularity = []
modularity_w = []

ds = np.arange(150,2200//2,50)
d = ds[16]

u = sort_corrs[d]
G = nx.from_pandas_adjacency(df>=u)
modulos = community_louvain.best_partition(G)
modularity_ = community_louvain.modularity(modulos,G)
modularity.append(modularity_)
set(modulos.values())
```

```python
#nodo_objetivo = 1
coeficiente_de_participacion = []
for nodo_objetivo in list(G.nodes()):
    degree_total_nodo_objetivo = G.degree(nodo_objetivo)
    clusters = list(set(modulos.values()))
#     print({"nodo_para_estudiar":nodo_objetivo,"cluster":modulos[nodo_objetivo], "degree_total_nodo_objetivo":degree_total_nodo_objetivo})
    clusters.remove(modulos[nodo_objetivo]) # remuevo el cluster al que pertenece de la lista
    suma = 0
    if degree_total_nodo_objetivo==0:
        continue
    for k in clusters:
        nodos_cluster_kesimo = [_nodo for _nodo, _community in modulos.items() if _community==k ]
        kesimo_degree_nodo_objetivo = G.subgraph([nodo_objetivo,*nodos_cluster_kesimo]).degree(nodo_objetivo)
#         print({"nodo":nodo_objetivo,"cluster":k, "degree intra-cluster":kesimo_degree_nodo_objetivo})
        suma+=(kesimo_degree_nodo_objetivo/degree_total_nodo_objetivo)**2
    print({"nodo_para_estudiar":nodo_objetivo,"cluster":modulos[nodo_objetivo], "degree_total":degree_total_nodo_objetivo, "participacion":round(suma,2)})
    coeficiente_de_participacion.append(suma)
```
